```python
from dt import DecisionTree
import pandas as pd
```

```python
df = pd.DataFrame({"a": range(10), "b":range(1,11)})
df["target"] = (df["a"] > 5) * 1 * (df["b"] % 2 == 0)
```

```python
df
```

```python
dt = DecisionTree()
dt.fit(df, ["a", "b"], "target")
```

```python
X = pd.DataFrame({"a": [10,12,5,6,9,0,1,5,100,2], "b":range(1,11)})
X
```

```python
dt.predict(X)
```
